This previous command is called a magic command and works only in jupyter notebooks. If we are not using jupyter notebooks then the only way to see this plots is to use  `plt.show()`

```python
x = np.arange(0,10)
```

```python
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x, y, 'red')
```

```python
plt.plot(x, y, 'r--')
plt.xlim(0,4)
plt.ylim(0,10)
plt.title('Title of the plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
```

Understand that in case of jupyter notebook, all these commands need to go in the same cell otherwise jupyter will not be able to figure out which commands go with which plots
